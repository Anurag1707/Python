```python
lst=[[1,3,5,6],[3,4,5],[3,5,7],[10,20,30]] 
ans=[]
for i in range(len(lst)):
    temp=lst[i].copy()
    if i%2==0:
        for j in range(len(temp)):
            if temp[j]%2==0:
                temp[j]-=1
        ans.append(temp)    
    else:
        for j in range(len(temp)):
            if temp[j]%2!=0:
                temp[j]+=1
        ans.append(temp)    
print(lst)
print(ans)
```
